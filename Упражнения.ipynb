```python
import time

# Выгрузим id пользователей всей группы

count = 1000

offset = 0

user_ids = []

while offset < data['response']['count']:

    print('Выгружаю {} пользователей с offset = {}'.format(count, offset))    

    params = {

        'group_id': 'love_run',

        'v': 5.73,

        'count': count,

        'offset': offset,
        
        'access_token': '2b38c20a2b38c20a2b38c20af62b5dcfa722b382b38c20a7003e8d2faf88db0dd1dddc0'

    }

    # такой же запрос как в прошлый раз

    r = requests.get(url, params = params)

    data_full = r.json() 

    user_ids += data_full['response']['items']

    # увеличиваем смещение на количество строк выгрузки

    offset += count

    print('Ожидаю 0.2 секунды...')

    time.sleep(0.2)
    
len (user_ids)
```

```python
#метод образения к апи вк дает информацию по 1000 пользователям максиму, разобьем наш лист на части по 1к

rows_limit = 1000

user_lists = []

position = 0

while position < len(user_ids):

    user_lists.append( user_ids[position:position + rows_limit] )

    position += rows_limit
    
# Тем самым получим лист user_lists из 58 листов по 1000 пользователей (для примера указаны порядковые номера 
# пользователей от 0 до 999):

print(user_lists[57])

#проветим крайний лист, в нем должно быть 256 пользователей

len(user_lists)
```

```python
#сделали проверку листов, теперь сделаем словарь с городами

city_stats = {}

for item in city_list:
    
    if item in city_stats:
        
        city_stats[item] += 1
   
    else:
        
        city_stats[item] = 1
    
print(city_stats)
```

```python
#упорядочим словарь по значеням ключей

list_d = list(city_stats.items())
list_d.sort()
list_d
```

```python
#упорядочим словарь по значеням пользователей по нарастанию

list_d = list(city_stats.items())
list_d.sort(key=lambda i: i[1])
for i in list_d:
     print(i[0], ':', i[1])
```

```python
#упорядочим словарь по значеням пользователей по убыванию

list_d = list(city_stats.items())
list_d.sort(key=lambda i: i[1], reverse=True)
for i in list_d:
     print(i[0], ':', i[1])
```
